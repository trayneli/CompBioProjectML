```python
EC_Descriptor = pd.read_csv("EC_Descriptor.tsv", sep = "\t", names = ["Gene", "Chr_Pos", "UniProtID", "Descriptor_1", "Descriptor_2"])
EC_Descriptor.head()
```

```python
Descriptor_col = [str(x) + str(y) if str(y) != "nan" else str(x) for x,y in zip(EC_Descriptor["Descriptor_1"].tolist(), EC_Descriptor["Descriptor_2"].tolist())]
Descriptor_col
```

```python
EC_Descriptor["Descriptor"] = Descriptor_col
EC_Descriptor = EC_Descriptor.drop(columns = ["Descriptor_1", "Descriptor_2"])
```

```python
EC_Descriptor.head()
```

```python
New_Descriptor = []
EC_List = []
for i in range(len(EC_Descriptor)):
    Descriptor_list = EC_Descriptor["Descriptor"][i].split(" ")
    Trimmed_Descriptor = []
    EC_List_str = ""
    j_iter = iter(range(len(Descriptor_list)))
    for j in j_iter:
        if (j < 2):
            continue
        elif (Descriptor_list[j] == "(EC"):
            EC_List_str += Descriptor_list[j][1:]
            EC_List_str += " "
            EC_List_str += Descriptor_list[j + 1][:-1]
            EC_List_str += "|"
            next(j_iter, None)
        else:
            Trimmed_Descriptor.append(Descriptor_list[j])
    Trimmed_Descriptor_str = " ".join(Trimmed_Descriptor)
    EC_List.append(EC_List_str)
    New_Descriptor.append(Trimmed_Descriptor_str)

EC_List


```

```python
for i in range(len(EC_List)):
    EC_str_list = EC_List[i].split("|")
    EC_str_list = EC_str_list[:-1]
    if (len(EC_str_list) > 0):
        for j in range(len(EC_str_list)):
            if "-" in EC_str_list[j]:
                continue
            elif (j == (len(EC_str_list) - 1)):
                EC_List[i] = EC_str_list[j]
            else:
                EC_List[i] = EC_str_list[j]

```

```python
EC_Descriptor["Descriptor"] = New_Descriptor
EC_Descriptor["EC"] = EC_List
```

```python
EC_Descriptor.head()
```

```python
EC_Descriptor.to_csv("Parsed_EC_Descriptor.tsv", index = False, sep = "\t")
```
