```python
import pandas as pd
import numpy as np
import subprocess
```

```python
fasta_files = subprocess.run("ls Sequences/ProteinData/*.fasta", shell = True, capture_output = True, text=True).stdout
```

```python
fasta_files_list = fasta_files.strip().split("\n")
print(fasta_files_list)
print(len(fasta_files_list))
```

```python
EC_Data = pd.read_csv("Annotation_EC_Data.tsv", sep="\t")
EC_Data.head()
```

```python
t = EC_Data[EC_Data["UniProtID"] == "A2PYH4"].reset_index()
t["EC"][0]
```

```python
sequences = []
IDs = []
ECs = []
for file in fasta_files_list:
    with open(file, 'r') as fin:
        data = fin.readlines()
        stripped_data = [x.strip() for x in data]
        sequence = "".join(stripped_data[1:])
        sequences.append(sequence)
    name = file.split("/")[-1]
    ID = name[:-6]
    IDs.append(ID)
    EC = EC_Data[EC_Data["UniProtID"] == ID].reset_index()
    EC_Num = EC["EC"][0]
    ECs.append(EC_Num)

print(sequences[0:10])
print(IDs[0:10])
print(ECs[0:10])
                      
                      
```

```python
Sequence_Frame = pd.DataFrame({"UniProtID": IDs, "Sequence" : sequences, "EC" : ECs})
Sequence_Frame.head()
```
